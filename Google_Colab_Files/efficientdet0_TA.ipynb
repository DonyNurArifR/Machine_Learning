```python
import numpy as np
import os

from tflite_model_maker.config import ExportFormat, QuantizationConfig
from tflite_model_maker import model_spec
from tflite_model_maker import object_detector

from tflite_support import metadata

import tensorflow as tf
assert tf.__version__.startswith('2')

tf.get_logger().setLevel('ERROR')
from absl import logging
logging.set_verbosity(logging.ERROR)
```

```python
from google.colab import drive
drive.mount('/content/gdrive')
```

```python
train_data = object_detector.DataLoader.from_pascal_voc('/content/gdrive/MyDrive/KULIAH PENS/Tugas Akhir/Tugas Akhir/dataset215_jpg_baru/train',
                                                        '/content/gdrive/MyDrive/KULIAH PENS/Tugas Akhir/Tugas Akhir/dataset215_jpg_baru/train',
                                                        ['glass', 'metal', 'paper', 'plastic'])
```

```python
val_data = object_detector.DataLoader.from_pascal_voc('/content/gdrive/MyDrive/KULIAH PENS/Tugas Akhir/Tugas Akhir/dataset215_jpg_baru/val',
                                                      '/content/gdrive/MyDrive/KULIAH PENS/Tugas Akhir/Tugas Akhir/dataset215_jpg_baru/val',
                                                      ['glass', 'metal', 'paper', 'plastic'])
```

```python
spec = model_spec.get('efficientdet_lite0')
```

```python
model = object_detector.create(train_data, model_spec=spec, batch_size=4, train_whole_model=True, epochs=100, validation_data=val_data)
```

```python
model.evaluate(val_data)
```

```python
model.export(export_dir='/content/gdrive/MyDrive/KULIAH PENS/Tugas Akhir/Tugas Akhir', tflite_filename='efficientdet_baru.tflite')
```

```python
model.evaluate_tflite('/content/gdrive/MyDrive/KULIAH PENS/Tugas Akhir/Tugas Akhir/efficientdet0.tflite', val_data)
```
